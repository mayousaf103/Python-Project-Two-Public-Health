2. DATA CLEANING/TIDYING

```python
#Replacing all the missing values of '?' with the properly designated numpy NaNs
#Created a list of missing values
missing_values = ['?']
df=pd.read_csv ('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv', na_values = missing_values)
```

```python
#Column Types
df.dtypes
```

```python
#Dropping columns which are irrelevant/not needed for the data analysis
#Those columns contain the 's', 'na', or very little information which is not
#helpful for the data analysis
drop_cols=['weight',
       'admission_type_id', 'discharge_disposition_id', 'admission_source_id',
      'payer_code', 'medical_specialty', 'num_procedures',
       'number_outpatient', 'number_inpatient', 'diag_1',
       'diag_2', 'diag_3', 'number_diagnoses', 'max_glu_serum',
       'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
       'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
       'tolazamide', 'examide', 'citoglipton',
       'glyburide-metformin', 'glipizide-metformin',
       'glimepiride-pioglitazone', 'metformin-rosiglitazone',
       'metformin-pioglitazone', 'A1Cresult']
df.drop(drop_cols, inplace=True, axis=1)
df = df.dropna()
df
```

```python
#Renaming columns
df.rename(columns={'race':'ethnicity'})
```

```python
#Missing values
df.isnull()
```

```python
df.isnull().sum()
```

3. EXAMINATION AND COMPARISON OF DATA

```python
#Creating a dataframe for the elderly (70-80)
elderly=['[70-80)']
```

```python
retired_elderly=df[df.age.isin(elderly)]
```

```python
#Confirming the dataframe
retired_elderly
```

```python
#Total of 26068 rows and 43 columns
retired_elderly.shape
```

```python
#Creating a second dataframe for young adults (20-30) to compare to the first dataframe
young_adults=['[20-30)']
```

```python
university_working_adults=df[df.age.isin(young_adults)]
```

```python
#Confirming the dataframe
university_working_adults
```

```python
#Total of 1657 rows and 43 columns
university_working_adults.shape
```

```python
#Will join both of the age groups to create a new dataframe
join = pd.concat([retired_elderly, university_working_adults], ignore_index=True)
```

```python
#Confirming the join dataframe
join
```

```python
#Total of 27725 rows and 43 columns
join.shape
```

4. MEASURES OF CENTRAL TENDENCY

```python
#Aggregating Data/Grouping
join.groupby(['age'])['encounter_id'].count()
```

```python
#To display the mean value by gender for both the elderly and young adults
join.groupby(['gender'])['patient_nbr'].mean()
```

```python
#To display the mean value by race for the young adults and elderly
join.groupby(['race'])['encounter_id'].mean()
```

```python
#To view the min value by gender 
join.groupby(['gender'])['time_in_hospital'].min()
```

```python
#To view the max value by race
join.groupby(['race'])['time_in_hospital'].max()
```

```python
#To see the median value for the time in hospital
join['time_in_hospital'].median()
```

```python
#To see the mode for the discharge disposition id
join['num_medications'].mode()
```

```python
#To see if the insulin column is unique or not
join['insulin'].is_unique
```

```python
#To see if the diabetesMed column is unique or not
join['diabetesMed'].is_unique
```

```python
#To display the count by patient nbr for the insulin for the elderly
retired_elderly.groupby(['insulin']) ['patient_nbr'].count()
```
